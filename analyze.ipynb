```python
df.columns=headers
```

```python
df.replace('?',np.nan,inplace=True)
print(df)
```

```python
missed=df.isnull()
```

```python
bo=df['bore'].astype('float').mean()
df.replace(np.nan,bo,inplace=True)

str=df['stroke'].astype('float').mean()
df.replace(np.nan,str,inplace=True)


hor=df['horsepower'].astype('float').mean()
df.replace(np.nan,hor,inplace=True)

peak=df['peak-rpm'].astype('float').mean()
df.replace(np.nan,norm,inplace=True)
df
```
